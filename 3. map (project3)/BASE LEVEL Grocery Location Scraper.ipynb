```python
# Dependencies
import pandas as pd
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
from prettyprinter import pprint
import json
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

### Grab the Giant Food Store Addresses for Just Alexandria, Virginia

```python
url = 'https://stores.giantfood.com/va/alexandria'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
print(soup.prettify())
```

```python
results = soup.find_all('li', class_="LocationList-item l-col-xs-12 l-col-sm-6 l-col-md-3-up")
print(f"stores = {len(results)}")
```

```python
title = soup.find_all('a', class_="Teaser-titleLink Link--primary")
print(f"titles = {len(title)}")
```

```python
street_address = soup.find_all('span', class_="c-address-street-1")
print(f"street addresses = {len(street_address)}")
```

```python
city = soup.find_all('span', class_="c-address-city")
print(f"city ids = {len(city)}")
```

```python
state = soup.find_all('abbr', class_="c-address-state")
print(f"state ids = {len(state)}")
```

```python
zip_code = soup.find_all('span', class_="c-address-postal-code")
print(f"zip codes = {len(zip_code)}")
```

```python
stores = results
```

```python
stores_list = []

for store in stores:

    try:
        street_address = store.find('span', class_='c-address-street-1').text
        city = store.find('span', class_='c-address-city').text
        state = store.find('abbr', class_='c-address-state').text
        zip_code = store.find('span', class_='c-address-postal-code').text

        product_dict = {
            'street_address': street_address,
            'city': city,
            'state': state,                
            'zip_code': zip_code
        }

        stores_list.append(product_dict)

    except Exception as e:
        print(e)
        pass
```

```python
stores_df = pd.DataFrame.from_dict(stores_list)
stores_df
```

### Grab the Giant Food Store States

```python
url = 'https://stores.giantfood.com'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
resultsA = soup.find_all('li', class_="DirectoryList-item")
print(f"states = {len(resultsA)}")
```

```python
resultsB = soup.find_all('a', class_="DirectoryList-itemLink Link--secondary")
print(f"states = {len(resultsB)}")
```

```python
resultsA[0].text
```

```python
store_states = resultsA
```

```python
states_list = []

for store_state in store_states:

    try:
        state = store_state.find('a', class_="DirectoryList-itemLink Link--secondary").text        
        
        state_dict = {            
            'state': state
        }

        states_list.append(state_dict)

    except Exception as e:
        print(e)
        pass
```

```python
states_df = pd.DataFrame.from_dict(states_list)
states_df
```

```python
state_abbr_dict = {
    "Washington DC": "dc",
    "Delaware": "de",
    "Maryland": "md",
    "Virginia": "va"
}
```

```python
state_abbr_dict
```

```python
stores_list = []

for store in stores:
    print(store)

    input_field = browser.find_by_id('typeahead-search-input')
    input_field.fill(store)
    search_button = browser.find_by_css('button.search-button')
    search_button.click()

    time.sleep(10)

    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')

    product_list = soup.find('ul', class_='product-list')
    products = product_list.find_all('li')

    for p in products:
        try:
            price = p.find('span', class_='product-grid-cell_main-price').text
            title = p.find('div', class_='product-tile_detail-title').h3.text
            size = p.find('span', class_='product-grid-cell_size').text
            unit = p.find('span', class_='product-grid-cell_unit').text

            product_dict = {
                'ingredient': store,
                'price': price,
                'title': title,
                'size': size,
                'unit': unit
            }

            stores_list.append(product_dict)

        except Exception as e:
            print(e)
            pass
```

```python
product_df = pd.DataFrame(stores_list)
pd.set_option('display.max_rows', None)
product_df
```
